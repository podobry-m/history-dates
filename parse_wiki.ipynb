```python
import requests
from bs4 import BeautifulSoup

```

```python
url = "https://ru.wikipedia.org/wiki/Хронология_истории_России"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Парсим даты и события (пример для Википедии)
for li in soup.select('li'):
    text = li.get_text()
    if '—' in text:  # Ищем строки с датами
        date, event = text.split('—', 1)
        print(date.strip(), event.strip())
```
